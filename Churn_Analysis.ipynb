```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

The goal of this analysis is to identify patterns of customer churn, determine key drivers of retention, and evaluate customer lifetime trends for a subscription-based business. Insights from this analysis can inform strategies to reduce churn and improve long-term revenue.Overall Churn



```python
# Count customers by gender
gender_counts = df["gender"].value_counts()

print("Customer Count by Gender:")
print(gender_counts)

# Total number of customers
total_customers = df.shape[0]
print("\nTotal Customers:", total_customers)


# Bar chart
plt.figure()

gender_counts.plot(kind="bar")

plt.title("Customer Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")

plt.show()
```

Gender distribution:

Male: 3,549 (50.4%)
Female: 3,483 (49.6%)

Observation: The gender distribution is nearly even, indicating that gender alone is unlikely to be a major factor in churn.

```python
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,72],
    labels=["0-12 Months","13-24 Months","25-48 Months","48+ Months"]
)

tenure_churn = pd.crosstab(df["TenureGroup"], df["Churn"], margins=True)

print(tenure_churn)

plt.figure()

pd.crosstab(df["TenureGroup"], df["Churn"]).plot(kind="bar")

plt.title("Churn by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Number of Customers")

plt.show()
```

Analysis:

Churn is highest among customers within the first year of subscription, with nearly 48% leaving.
Churn decreases significantly for customers with tenure longer than 24 months, showing that once a customer stays past the initial period, loyalty increases.
Insight: Early-stage retention programs, such as onboarding guidance, personalized offers, and engagement campaigns, could reduce churn during the first year.

```python
plt.figure()

sns.countplot(x="Churn", data=df)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

Churned: 1,869 (26.5%)

```python
contract_churn = pd.crosstab(df["Contract"], df["Churn"], margins=True)

print("Contract Type vs Churn")
print(contract_churn)
plt.figure()

sns.countplot(x="Contract", hue="Churn", data=df)

plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.show()
```

Customers on month-to-month contracts have the highest churn rate (~42.7%), highlighting that short-term commitments increase the likelihood of leaving.
Customers with one-year contracts have a churn rate of ~11%, and two-year contracts ~2.8%, indicating that longer-term contracts strongly improve retention.
Insight: Encouraging customers to commit to longer-term contracts could substantially reduce churn. Marketing strategies could include discounts or incentives for long-term contract adoption.

```python

# Count customers by Internet Service
internet_counts = df["InternetService"].value_counts()

print("Customer Count by InternetService Status:")
print(internet_counts)

# Total number of customers
total_customers = df.shape[0]
print("\nTotal Customers:", total_customers)

# Crosstab showing InternetService vs Churn
internet_churn = pd.crosstab(df["InternetService"], df["Churn"], margins=True)

print("\nInternet Service vs Churn:")
print(internet_churn)

# Graph showing churn distribution
plt.figure()

sns.countplot(x="InternetService", hue="Churn", data=df)

plt.title("Customer Churn by Internet Service Type")
plt.xlabel("Internet Service Type")
plt.ylabel("Number of Customers")

plt.show()
```

Analysis:

Fiber optic users have the highest churn rate (41.9%), likely due to higher expectations or pricing sensitivity.
DSL and customers with no internet service show lower churn rates (19% and 7%, respectively).
Insight: Customer experience, service quality, and perceived value are crucial for high-speed internet users. Consider loyalty programs, premium support, or service guarantees to reduce churn in this segment.

```python
# Create tenure groups
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0,12,24,48,72],
    labels=["0-12 Months","13-24 Months","25-48 Months","48+ Months"]
)

# Crosstab with churn
tenure_churn = pd.crosstab(df["TenureGroup"], df["Churn"], margins=True)

print("Tenure Group vs Churn")
print(tenure_churn)
plt.figure()

sns.countplot(x="TenureGroup", hue="Churn", data=df)

plt.title("Customer Churn by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Number of Customers")

plt.show()
```

Churn is highest in the first year (0–12 months), almost 48% of customers churned in this range.
Customers staying longer than 2 years are much more loyal, indicating tenure is a key retention driver.

```python
monthly_stats = df.groupby("Churn")["MonthlyCharges"].agg(
    Number_of_Customers="count",
    Average_Monthly_Charges="mean").round(2)

print(monthly_stats)
plt.figure()

sns.boxplot(x="Churn", y="MonthlyCharges", data=df)

plt.title("Monthly Charges vs Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

Customers who churn pay higher average monthly charges of 74.44 compared to
the lower average of 61.27, indicating that price is a significant factor in leaving the service.
High-paying customers may expect more value from the service. A lack of perceived value could lead to churn.

**Insight:** Consider offering additional services, discounts, or loyalty incentives for high-paying customers to reduce churn.

```python


# Create a table showing Payment Method vs Churn
payment_churn = pd.crosstab(df["PaymentMethod"], df["Churn"], margins=True)

# Rename the total column for clarity
payment_churn.rename(columns={"All": "Total"}, inplace=True)

print("Payment Method vs Churn Table:")
print(payment_churn)
plt.figure()

sns.countplot(x="PaymentMethod", hue="Churn", data=df)

plt.title("Churn by Payment Method")
plt.xticks(rotation=45)

plt.show()
```

**Analysis:**

Customers paying via electronic check have the highest churn (45%), while those on automatic payments (bank or credit card) have much lower churn (16%–17%).
Insight: Automating payments improves retention, likely because customers on automatic payments have fewer interruptions and feel more committed. Marketing campaigns could promote automatic payment adoption.

```python
numeric_df = df[["tenure","MonthlyCharges","TotalCharges"]]

plt.figure()

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Between Numerical Features")

plt.show()
```

**Key Retention Drivers**

Based on the analysis, the strongest drivers of retention include:

Contract Length: Longer-term contracts reduce churn substantially.

Customer Tenure: Retention increases with tenure; early-stage retention programs are critical.

Payment Method: Automatic payments (bank transfer or credit card) improve retention.

Service Type: Fiber optic users are more likely to churn; enhanced support and incentives could help.

Pricing: Higher monthly charges correlate with higher churn, highlighting the need for perceived value.

**Customer Lifetime Trends**

High-risk customers:

*   Month-to-month subscribers
*   Customers within their first 12 months
*   Fiber optic internet users
*   Electronic check payers
*   Customers with high monthly charges

Low-risk customers (loyal):

*   48+ months tenure
*   1-year or 2-year contracts
*   Fiber optic internet users
*   Automatic payment methods (bank or credit card)
*  Moderate monthly charges

**Business Implication:**

Target retention initiatives towards new, high-paying, and month-to-month customers. Incentivizing longer contracts, automatic payments, and better service experience can significantly reduce churn and increase revenue stability.

**Final Remarks:**

Churn is high in the early stages of the customer lifecycle.
Contract length, payment method, service type, and monthly charges are strong retention drivers.
Retention programs should focus on onboarding, engagement, and perceived value.
